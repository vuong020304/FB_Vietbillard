```python
import numpy as np
```

```python
post_list = np.load('../Data/officialfoden.npy', allow_pickle=True)
post_list = post_list.tolist()
commenter_url = []
for post in post_list:
    if len(post['comments_full'])>0:
        for url in post['comments_full']:
            s = url['commenter_url']
            s = s[21:]
            print(s)
            commenter_url.append(s)
            
print('Số người bình luận:', len(commenter_url))
```

```python
commenter_url = []
for post in post_list:
    if len(post['comments_full'])>0:
        for url in post['comments_full']:
            s = url['commenter_id']
            print(s)
            commenter_url.append(s)
            
print('Số người bình luận:', len(commenter_url))
```
